```python
# This gets the index of the highest count (the mode)
df_land1 = df_land.groupby('osm_land_id', as_index=False).agg({
    'area_2':'sum',
    'ratio': 'sum',
    'fclass': lambda x: x.value_counts().index[0]
})
df_land1.rename(columns={'osm_land_id':'osm_id','area_2':'total_land_area','ratio':'total_land_ratio','fclass':'major_land_type'}, inplace= True)
```

##### Final score

```python
df_box.rename(columns={'osm_box_id':'osm_id'}, inplace =True)
```

```python
import pandas as pd
from functools import reduce

# 1. List your dataframes
dfs_to_merge = [
    df_box, df_land1, df_dso1, df_dsomocy1, df_road1, 
    df_railway1, df_plant1, df_dem1, df_dni1, df_temp1, df_pvout1
]

# 2. Clean the "right" dataframes (ensure 1 row per osm_id)
# We keep df_box as is, but drop duplicates from all others
cleaned_dfs = [dfs_to_merge[0]] + [
    df.drop_duplicates(subset='osm_id', keep='first') for df in dfs_to_merge[1:]
]

# 3. Perform the merge
final_df = reduce(
    lambda left, right: pd.merge(left, right, on='osm_id', how='left'), 
    cleaned_dfs
)

print(f"Final row count: {len(final_df)}")
```
